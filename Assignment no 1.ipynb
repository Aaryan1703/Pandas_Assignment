## filling missing value

```python
for i in df2.columns:
    if(df2[i].dtypes=="object"):
        md=df2[i].mode()
        df2[i]=df2[i].fillna(md)
    elif(df2[i].dtypes!="object"):
        me=df2[i].mean()
        df2[i]=df2[i].fillna(me)
        
```

```python
df2.isnull().sum()
```

## Que 1: Find out car having highest Mileage in city

```python
max_milage=df2["MPG.city"].max()
max_milage
```

```python
df2[df2["MPG.city"]==max_milage]
```
